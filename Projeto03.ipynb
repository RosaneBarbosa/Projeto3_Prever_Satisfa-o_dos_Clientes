## 3. Pré-processamento dos Dados

### 3.1 Análise de Componentes Principais (PCA)

```python
# Importando os módulos para normalizar os dados e em seguida aplicar Análise de Componentes Principais (PCA)
# a fim de reduzir a grande dimensionalidade do arquivo.

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
# Criando um array e separando o array em input (excluindo a variável ID) e output
array = dados.values
X = array[:, 1:370]
Y = array[:, 370]
```

```python
# Primeiro temos que normalizar os dados antes de aplicar o PCA
scaler = MinMaxScaler(feature_range = (0, 1))
rescaledX = scaler.fit_transform(X)
```

```python
# Criando e treinando PCA para reduzir a dimensionalidade de 370 atributos para 10 componentes
pca = PCA(n_components = 10)
fitX = pca.fit(rescaledX)
```

```python
## Print da variância total dos dados expressada por cada componente
print("Variância: %s" % fitX.explained_variance_ratio_)
print("\nTotal de variância explicada pelos componentes principais: %.4f%%" % fitX.explained_variance_ratio_.sum())
```

```python
# Gráfico da variância dos dados expressada por cada componente
import pandas as pd
df_var = pd.DataFrame(fitX.explained_variance_ratio_.T)

df_var.plot(kind = 'bar', title = "Quantidade da variância total dos dados expressada por cada componente principal",
           legend = False, figsize = (10,5), color = 'purple')
plt.xlabel('Componente Principal')
plt.ylabel('Quantidade de Variância Explicada')
plt.show()
```

```python
# Aplicando PCA aos dados
X_PCA = pca.transform(rescaledX)
```

```python
# Shape após PCA
X_PCA.shape
```

### 3.2 Divisão dos Dados em Treino e Validação - Amostragem

```python
# Importando módulo do sklearn
from sklearn.model_selection import train_test_split
```

```python
# Definindo o tamanho das amostras e da semente (garantir a reprodução dos resultados)
teste_size = 0.33
seed = 7
```

```python
# Criando os conjuntos de dados de treino e de validação
X_treino, X_valida, Y_treino, Y_valida = train_test_split(X_PCA, Y, test_size = teste_size, random_state = seed)
```

### 3.3 SMOTE (Synthetic Minority Over-sampling Technique)

```python
# Importando a função SMOTE para balancear a variável target. A função SMOTE consiste em gerar dados sintéticos
# (não duplicados) da classe com menor quantidade
from imblearn.over_sampling import SMOTE
```

```python
# Aplicando a função SMOTE a fim de balancear os dados de treino
smt = SMOTE(random_state = seed)
X_treino_smt, Y_treino_smt = smt.fit_sample(X_treino, Y_treino)
```

```python
# Gráfico da distribuição da variável target após a aplicação da técnica de oversampling com SMOTE aos dados de treino.
sns.countplot(Y_treino_smt)
```

## 5. Seleção do Modelo Preditivo

```python
# Preparando a lista de modelos
modelos = []
modelos.append(('LR', LogisticRegression()))
modelos.append(('LDA', LinearDiscriminantAnalysis()))
modelos.append(('KNN', KNeighborsClassifier()))
modelos.append(('NB', GaussianNB()))
modelos.append(('CART', DecisionTreeClassifier()))
```

```python
# Loop com as métricas de avaliação de cada modelo
resultados_acuracia = []
resultados_auc = []
nomes = []

for nome, modelo in modelos:
    resultado_treino = modelo.fit(X_treino_smt, Y_treino_smt)
    Y_pred_valida = modelo.predict(X_valida)
    pred_valida = [round(value) for value in Y_pred_valida]
    acuracia = accuracy_score(Y_valida, pred_valida)
    resultados_acuracia.append(acuracia)
    auc = roc_auc_score(Y_valida, pred_valida)
    resultados_auc.append(auc) 
    nomes.append(nome)
    msg = "%s: \n Acurácia: %f%% \n AUC: %f \n" % (nome, (acuracia * 100), auc)
    print(msg) 
```

### Análise dos Modelos:
#### O modelo preditivo de Regressão Logística apresentou o melhor score para AUC embora a acurácia deste modelo de 70% seja menor do que a dos algoritmos CART e KNN.
#### Além disso, a Regressão Logística tem um menor erro do tipo II (quantidade de falsos negativos), neste caso é pior classificar errado o cliente como satisfeito (TARGET = 0) quando o cliente está insatisfeito (TARGET = 1).

## 6.2 Otimizando o Modelo - Ajuste de Hyperparâmetros
#### Usando Grid Search Parameter Tuning para encontrar os melhores hyperparametros a fim de ajustar o algoritmo XGBoost que apresentou a melhor performance (maior AUC e Acurácia) dentre todos os algoritmos validados.

```python
# Importando o método Grid Search Parameter
from sklearn.model_selection import GridSearchCV
```

```python
# Definindo os valores dos parâmetros a serem testados
# max_depth controla a profundidade, quanto maior a profundidade, mais complexo o modelo e maiores chances de overfitting.
# Gamma controla a regularização (ou evita overfitting), quanto maior o valor, maior a regularização.

valores_grid = {'max_depth': [4, 5, 6], 'gamma': [ 0.0, 0.5, 1.0]}
```

```python
# Criando o grid
grid = GridSearchCV(estimator = modelo, param_grid = valores_grid)
grid.fit(X_treino_smt, Y_treino_smt, eval_metric = "auc", early_stopping_rounds = 10, eval_set = [(X_valida, Y_valida)])
```

```python
# Print do resultado
print("AUC: %.3f" % (grid.best_score_))
print("Melhores Parâmetros do Modelo:\n", grid.best_estimator_)
```

```python
# XGBoost utilizando os melhores parâmetros
selected_model = grid.best_estimator_
selected_model.fit(X_treino_smt, Y_treino_smt)
```

```python
# Previsão utilizando os melhores parâmetros com os dados de validação
Y_pred_valida = selected_model.predict(X_valida)
pred_valida = [round(value) for value in Y_pred_valida]
```

```python
# Print das métricas
acuracia = accuracy_score(Y_valida, pred_valida)
print("Acurácia: %.3f%%" % (acuracia * 100))
auc = roc_auc_score(Y_valida, pred_valida)
print("AUC: %.3f" % auc)
```

```python
# Confusion Matrix
matrix = confusion_matrix(Y_valida, pred_valida)
print(matrix)
tn, fp, fn, tp = confusion_matrix(Y_valida, pred_valida).ravel()
(tn, fp, fn, tp)
```

### Conclusão:
#### O modelo preditivo com o algoritmo XGBoost apresentou um melhor desempenho do que o algoritmo de Regressão Logística para a métrica de AUC (maior score) bem como uma melhor acurácia.
#### Ajustando os hiperparâmetros do XGBoost com o método Grid Search Parameter, obtivemos uma performance melhor (maior AUC) e um erro do tipo II (Falso Negativo) menor que a Regressão Logística.
